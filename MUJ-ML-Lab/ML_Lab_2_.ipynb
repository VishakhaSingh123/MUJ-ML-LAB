```python
import pandas as pd

# Column names
cols = ["Age","Gender","Total_Bilirubin","Direct_Bilirubin",
        "Alkaline_Phosphotase","SGPT","SGOT",
        "Total_Proteins","Albumin","A_G_Ratio","Dataset"]

# Load dataset
df = pd.read_csv("../datasets/ILPD.csv", names=cols)

print("First 5 rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nShape:")
print(df.shape)
```

```python
num_cols = df.select_dtypes(include=['float64','int64'])

print("Mean:\n", num_cols.mean())

print("\nMedian:\n", num_cols.median())

print("\nMode:\n", num_cols.mode())
```

```python
print("Minimum values:\n", num_cols.min())

print("\nMaximum values:\n", num_cols.max())

print("\nSum:\n", num_cols.sum())

print("\nVariance:\n", num_cols.var())

print("\nStandard Deviation:\n", num_cols.std())
```

```python
Q1 = num_cols.quantile(0.25)
Q2 = num_cols.quantile(0.50)
Q3 = num_cols.quantile(0.75)

print("First Quartile (Q1):\n", Q1)
print("\nSecond Quartile (Median):\n", Q2)
print("\nThird Quartile (Q3):\n", Q3)
```

```python
corr = num_cols.corr()
print("Correlation Matrix:\n", corr)

cov = num_cols.cov()
print("\nCovariance Matrix:\n", cov)
```

```python
import matplotlib.pyplot as plt

num_cols.hist(figsize=(10,8))
plt.suptitle("Distribution of Liver Dataset Features")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

cols = ["Age","Gender","Total_Bilirubin","Direct_Bilirubin",
        "Alkaline_Phosphotase","SGPT","SGOT",
        "Total_Proteins","Albumin","A_G_Ratio","Dataset"]

df = pd.read_csv("../datasets/ILPD.csv", sep=r"\s+", names=cols)

# convert numeric columns
for col in df.columns:
    if col != "Gender":
        df[col] = pd.to_numeric(df[col], errors="coerce")

df = df.dropna()

numeric_cols = df.drop("Gender", axis=1)

numeric_cols.plot(kind="box", figsize=(10,6))
plt.title("Boxplot of Liver Dataset")
plt.show()
```
